```python
def get_bio():# Get bio
    try:
        bio = driver.find_element(by=By.XPATH, value="/html/body/div[1]/div/div[1]/div/div[1]/div/div/div[1]/div[1]/section/main/div/div[1]/div").text
    except:
        bio = ''
    if not bio == '':
        return bio
    elif bio == 'The link you followed may be broken, or the page may have been removed. Go back to Instagram.':
        return 'banned'
    else:
        return False
```

```python
def get_last_post():
    try:
        WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.CLASS_NAME, "_aagu"))).click()
        time.sleep(1)
        d = driver.find_element(by=By.CLASS_NAME, value="_aaqe")
        date = d.get_attribute('datetime')
        driver.back()
    except:
        date = False
    return date
```

```python
def get_followers_count():
    followers = driver.find_element(by=By.XPATH, value='/html/body/div[1]/div/div[1]/div/div[1]/div/div/div[1]/div[1]/section/main/div/ul/li[2]/a/div/span').get_property('title')
    return followers
```
