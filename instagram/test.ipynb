```python
def get_bio():# Get bio
    try:
        bio = driver.find_element(by=By.XPATH, value="/html/body/div[1]/div/div[1]/div/div[1]/div/div/div[1]/div[1]/section/main/div/div[1]/div").text
    except:
        bio = ''
    if not bio == '':
        return bio
    else:
        return False
```

```python
def get_last_post():
    try:
        WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.CLASS_NAME, "_aagu"))).click()
        time.sleep(1)
        d = driver.find_element(by=By.CLASS_NAME, value="_aaqe")
        date = d.get_attribute('datetime')
        driver.back()
    except:
        date = False
    return date
```

```python
def get_users():
    myclient = MongoClient("mongodb://localhost:27017/")
    # linking the datab
    db = myclient['Instagram']
    # linking the collections
    tweet_collection = db['public']
    tweet = tweet_collection.find({},{"_id": 0})
    return tweet
```

```python
import os
from pymongo import MongoClient
from pymongo import InsertOne

myclient = MongoClient('mongodb://localhost:27017')
# Database
db = myclient['Instagram']
# Collection name
collection = db['public_tot']
# Opening and uploading the json file.
requesting = list()
with open('public_tot.json', 'r', encoding='utf-8') as file:
    for line in file:
        my_dict = json.loads(line)
        requesting.append(InsertOne(my_dict))
    collection.bulk_write(requesting)
    myclient.close()
```
